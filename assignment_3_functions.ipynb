**1. What is the difference between a function and a method in Python?**

**Answer:**

In Python, both functions and methods are blocks of reusable code. The key difference lies in their association with objects.

  * **Function:** A function is a standalone block of code that is not associated with any particular object or class. You define it independently, and you can call it from anywhere in your program.
  * **Method:** A method is a function that is associated with an object or a class. It is defined within a class, and it operates on the data (attributes) of the object it belongs to. When you call a method, you call it on an instance of a class.

**Example:**

```python
# Function
def greet_function(name):
    return f"Hello, {name}!"

print(greet_function("Alice")) # Output: Hello, Alice!

# Method
class Dog:
    def __init__(self, name):
        self.name = name

    def bark(self): # bark is a method of the Dog class
        return f"{self.name} says Woof!"

my_dog = Dog("Buddy")
print(my_dog.bark()) # Output: Buddy says Woof!
```

**4. What is the purpose of the `return` statement in a Python function?**

**Answer:**

The `return` statement in a Python function serves two primary purposes:

1.  **Exiting the Function:** It immediately terminates the execution of the function. Any code after the `return` statement within the function will not be executed.
2.  **Sending a Value Back:** It sends a value (or multiple values) back to the caller of the function. If no value is specified with `return`, or if there's no `return` statement at all, the function implicitly returns `None`.

**Example:**

```python
def multiply(a, b):
    result = a * b
    return result # Returns the calculated result

def say_hello():
    print("Hello!")
    # No return statement, so it implicitly returns None

x = multiply(4, 5)
print(x) # Output: 20

y = say_hello()
print(y) # Output: Hello! (then) None
```

**5. What are iterators in Python and how do they differ from iterables?**

**Answer:**

  * **Iterable:** An object that can be iterated over, meaning it can return its members one at a time. Examples include lists, tuples, strings, and dictionaries. An object is iterable if it implements the `__iter__()` method, which returns an iterator.
  * **Iterator:** An object that represents a stream of data. It is responsible for returning the next item from the sequence. An iterator implements two methods:
      * `__iter__()`: Returns the iterator object itself.
      * `__next__()`: Returns the next item from the sequence. If there are no more items, it raises a `StopIteration` exception.

**Difference:**

An **iterable** is something you can loop over (e.g., with a `for` loop). An **iterator** is the actual object that keeps track of the current position during iteration and knows how to get the *next* item. You get an iterator from an iterable.

**Example:**

```python
# Iterable (a list)
my_list = [10, 20, 30]

# Getting an iterator from the iterable
my_iterator = iter(my_list)

# Using the iterator
print(next(my_iterator)) # Output: 10
print(next(my_iterator)) # Output: 20
print(next(my_iterator)) # Output: 30

try:
    print(next(my_iterator))
except StopIteration:
    print("No more items in the iterator.")
```

**6. Explain the concept of generators in Python and how they are defined.**

**Answer:**

**Generators** are a simple and memory-efficient way to create iterators. They are functions that, instead of returning a single value and exiting, "yield" a sequence of values one at a time, pausing their execution state between each `yield`. This makes them ideal for working with large datasets where loading everything into memory at once would be inefficient or impossible.

**How they are defined:**

Generators are defined like regular functions, but instead of using the `return` keyword to send back a result, they use the `yield` keyword. When a `yield` statement is encountered, the generator pauses, sends the value to the caller, and saves its internal state. The next time `next()` is called on the generator, it resumes execution from where it left off.

**Example:**

```python
def countdown(n):
    print("Starting countdown...")
    while n > 0:
        yield n
        n -= 1
    print("Countdown finished!")

# Creating a generator object
counter = countdown(3)

print(next(counter)) # Output: Starting countdown... (then) 3
print(next(counter)) # Output: 2
print(next(counter)) # Output: 1

try:
    print(next(counter))
except StopIteration:
    print("Generator exhausted.")
```

**7. What are the advantages of using generators over regular functions?**

**Answer:**

The primary advantages of using generators over regular functions, especially for generating sequences, are:

1.  **Memory Efficiency (Lazy Evaluation):** Generators produce items one by one and on demand (lazy evaluation). They don't store the entire sequence in memory. Regular functions often compute and return the entire result at once, which can consume a lot of memory for large sequences.
2.  **Performance:** Because they don't have to build the entire sequence in memory before returning, generators can be faster for large datasets, especially when only a portion of the sequence is needed.
3.  **Infinite Sequences:** Generators can represent infinite sequences since they don't need to generate all values upfront. Regular functions cannot practically return an infinite sequence.
4.  **Simplicity in Code:** The code for generators often looks cleaner and more concise than writing a custom iterator class with `__iter__` and `__next__` methods.

**Example:**

Consider generating a large sequence of numbers:

```python
# Regular function (less memory efficient for very large N)
def generate_list_of_squares(n):
    squares = []
    for i in range(n):
        squares.append(i * i)
    return squares

# Generator (memory efficient)
def generate_squares_generator(n):
    for i in range(n):
        yield i * i

# Using the generator (only generates squares as needed)
for square in generate_squares_generator(1_000_000):
    if square > 100:
        break # We only needed a few squares, so we stop early
print(square)
```

**8. What is a lambda function in Python and when is it typically used?**

**Answer:**

A **lambda function** (also known as an anonymous function) is a small, single-expression function that you don't explicitly define with the `def` keyword. It's concise and is often used for short, simple operations.

**Syntax:** `lambda arguments: expression`

**When it is typically used:**

Lambda functions are typically used in situations where:

1.  **A small, throwaway function is needed for a short period:** They are useful for functions that are not going to be reused.
2.  **As arguments to higher-order functions:** Functions like `map()`, `filter()`, `sorted()`, and `reduce()` often accept a function as an argument, and lambda functions are perfect for providing a simple custom logic on the fly.
3.  **For simple operations that can be expressed in a single line:** If the logic is more complex, a regular `def` function is usually preferred for readability.

**Example:**

```python
# Using lambda with sorted()
students = [('Alice', 25), ('Bob', 20), ('Charlie', 30)]
# Sort by age
sorted_students = sorted(students, key=lambda student: student[1])
print(sorted_students) # Output: [('Bob', 20), ('Alice', 25), ('Charlie', 30)]

# Simple lambda function for adding two numbers
add = lambda x, y: x + y
print(add(10, 20)) # Output: 30
```

**9. Explain the purpose and usage of the `map()` function in Python.**

**Answer:**

The `map()` function in Python is a built-in higher-order function that applies a given function to each item of an iterable (like a list, tuple, etc.) and returns an iterator that yields the results. It's a concise way to transform elements in an iterable.

**Purpose:** To apply a transformation to every item in an iterable.

**Usage:**

The syntax is `map(function, iterable, ...)`.

  * `function`: The function to which each item of the iterable will be passed.
  * `iterable`: One or more iterables whose elements will be passed to the function.

The `map()` function returns a `map` object (which is an iterator). You often convert this to a list or tuple to see the results.

**Example:**

```python
numbers = [1, 2, 3, 4, 5]

# Using map to square each number
squared_numbers_map = map(lambda x: x * x, numbers)
squared_numbers_list = list(squared_numbers_map)
print(squared_numbers_list) # Output: [1, 4, 9, 16, 25]

# Using map with a regular function to convert to string
def to_string(num):
    return str(num)

string_numbers_map = map(to_string, numbers)
string_numbers_list = list(string_numbers_map)
print(string_numbers_list) # Output: ['1', '2', '3', '4', '5']
```

**10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

**Answer:**

These are three common higher-order functions in Python, each serving a distinct purpose for processing iterables:

1.  **`map()`:**

      * **Purpose:** To **transform** each element of an iterable. It applies a function to every item and returns a new iterable (an iterator) containing the results.
      * **Output:** An iterable with the same number of elements as the input, but with each element transformed.
      * **Analogy:** One-to-one transformation. Like applying a mold to each cookie dough.

2.  **`filter()`:**

      * **Purpose:** To **select** elements from an iterable that satisfy a certain condition. It applies a function (which should return a boolean) to each item and returns a new iterable (an iterator) containing only the items for which the function returned `True`.
      * **Output:** An iterable with potentially fewer elements than the input, containing only the filtered items.
      * **Analogy:** Filtering out unwanted items. Like a sieve.

3.  **`reduce()`:** (Note: `reduce` is in the `functools` module in Python 3, not a built-in function)

      * **Purpose:** To **aggregate** or **reduce** an iterable to a single cumulative result. It applies a function (that takes two arguments) cumulatively to the items of an iterable, from left to right, to reduce the iterable to a single value.
      * **Output:** A single value.
      * **Analogy:** Combining all items into one. Like crumbling all cookies into one bowl.

**Example:**

```python
from functools import reduce

numbers = [1, 2, 3, 4, 5]

# map(): Square each number
squared = list(map(lambda x: x * x, numbers))
print(f"map() result (squared): {squared}") # Output: [1, 4, 9, 16, 25]

# filter(): Keep only even numbers
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(f"filter() result (evens): {evens}") # Output: [2, 4]

# reduce(): Sum all numbers
total_sum = reduce(lambda x, y: x + y, numbers)
print(f"reduce() result (sum): {total_sum}") # Output: 15
```

**1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.**

```python
def calculate_even_sum(numbers_list):
    total_even = 0
    for num in numbers_list:
        if isinstance(num, (int, float)) and num % 2 == 0:
            total_even += num
    return total_even

kohli_scores = [45, 82, 33, 90, 12, 57, 66]
pujara_scores = [10, 25, 41, 60]
empty_list = []

print(f"Sum of even scores for Kohli: {calculate_even_sum(kohli_scores)}")
print(f"Sum of even scores for Pujara: {calculate_even_sum(pujara_scores)}")
print(f"Sum of even numbers in empty list: {calculate_even_sum(empty_list)}")
```

**4. Write a Python function that checks if a given number is prime or not from 1 to 200.**

```python
def check_prime(number):
    if not isinstance(number, int):
        return False
    if number <= 1:
        return False
    if number == 2:
        return True
    if number % 2 == 0:
        return False

    i = 3
    while i * i <= number:
        if number % i == 0:
            return False
        i += 2
    return True

print("Prime numbers from 1 to 200:")
prime_numbers_upto_200 = [n for n in range(1, 201) if check_prime(n)]
print(prime_numbers_upto_200)

print(f"Is 23 prime? {check_prime(23)}")
print(f"Is 100 prime? {check_prime(100)}")
print(f"Is 139 prime? {check_prime(139)}")
```

**5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.**

```python
class IndianFibonacciSeries:
    def __init__(self, limit):
        if not isinstance(limit, int) or limit < 0:
            raise ValueError("Limit must be a non-negative integer.")
        self.limit = limit
        self.counter = 0
        self.first = 0
        self.second = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.counter < self.limit:
            if self.counter == 0:
                self.counter += 1
                return self.first
            elif self.counter == 1:
                self.counter += 1
                return self.second
            else:
                next_val = self.first + self.second
                self.first = self.second
                self.second = next_val
                self.counter += 1
                return next_val
        else:
            raise StopIteration

print("Fibonacci sequence for 8 terms:")
for fib_num in IndianFibonacciSeries(8):
    print(fib_num, end=" ")
print("\n")

print("Fibonacci sequence for 3 terms:")
print(list(IndianFibonacciSeries(3)))
```

**6. Write a generator function in Python that yields the powers of 2 up to a given exponent.**

```python
def power_of_two_generator(max_exponent):
    if not isinstance(max_exponent, int) or max_exponent < 0:
        raise ValueError("Maximum exponent must be a non-negative integer.")

    for i in range(max_exponent + 1):
        yield 2**i

print("Powers of 2 up to exponent 6:")
for val in power_of_two_generator(6):
    print(val, end=" ")
print("\n")

my_gen = power_of_two_generator(3)
print(next(my_gen))
print(next(my_gen))
```

**7. Implement a generator function that reads a file line by line and yields each line as a string.**

```python
import os

def read_file_by_lines(file_path):
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"The file '{file_path}' was not found.")

    try:
        with open(file_path, 'r', encoding='utf-8') as f_obj:
            for line_data in f_obj:
                yield line_data
    except Exception as e:
        print(f"An issue occurred: {e}")

file_name_txt = "mahabharat_lines.txt"
sample_content = """Arjuna's Bow.
Krishna's Wisdom.
Draupadi's Resolve.
Dharma wins."""

with open(file_name_txt, "w") as f:
    f.write(sample_content)

print(f"Content of '{file_name_txt}':")
for line_read in read_file_by_lines(file_name_txt):
    print(line_read.strip())

os.remove(file_name_txt)

try:
    for line in read_file_by_lines("non_existent.txt"):
        pass
except FileNotFoundError as e:
    print(f"\nError encountered: {e}")
```

**9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.**

```python
def convert_celsius_to_fahrenheit(celsius_temp):
    return (celsius_temp * 9/5) + 32

mumbai_temperatures_c = [28.5, 32.0, 26.1, 30.7, 29.3]

print(f"Mumbai temperatures in Celsius: {mumbai_temperatures_c}")

mumbai_temperatures_f = list(map(convert_celsius_to_fahrenheit, mumbai_temperatures_c))

print(f"Mumbai temperatures in Fahrenheit: {mumbai_temperatures_f}")

delhi_temps_c = [25, 38, 42]
delhi_temps_f = list(map(lambda c: (c * 9/5) + 32, delhi_temps_c))
print(f"Delhi temperatures in Fahrenheit: {delhi_temps_f}")
```

**10. Create a Python program that uses `filter()` to remove all the vowels from a given string.**

```python
def filter_vowels_from_string(input_str):
    vowel_chars = "aeiouAEIOU"
    filter_logic = lambda char: char not in vowel_chars

    filtered_list = filter(filter_logic, input_str)
    result_str = "".join(filtered_list)
    return result_str

poem_line = "The quick brown fox jumps over the lazy dog"
indian_phrase = "Bharatham is my country"

print(f"Original: '{poem_line}'")
print(f"Without vowels: '{filter_vowels_from_string(poem_line)}'")

print(f"\nOriginal: '{indian_phrase}'")
print(f"Without vowels: '{filter_vowels_from_string(indian_phrase)}'")
```
